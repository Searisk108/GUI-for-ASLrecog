```python
import tkinter as tk
from tkinter import ttk # To style the widgets
from PIL import ImageTk
from PIL import Image, ImageDraw, ImageFont
import cv2
import IPython.display


def capture_and_save():
    cap = cv2.VideoCapture(0)

    if cap.isOpened(): # try to get the first frame
        ret, PersonView = cap.read()
    else:
        ret = False

    while ret:   

        ret, PersonView = cap.read()
        PersonView = cv2.flip(PersonView,1)


        #start_point (x1,y1) and end_point (x2,y2)
        y1 = 70
        y2 = 270
        x1 = 400
        x2 = 600

        HandView = PersonView[y1-1:y2+1, x1-1:x2+1] 
        cv2.rectangle(PersonView,(x1,y1),(x2,y2),(255,0,0),1) 
        
        #HandView = cv2.cvtColor(HandView, cv2.COLOR_BGR2GRAY)
        #HandView = cv2.GaussianBlur(HandView, (5,5), 0)

        cv2.imshow("Hand View", HandView)


        k = cv2.waitKey(5) & 0xFF
        if k == 32: #press space to capture photo
            cv2.imwrite("C:/Python/Python37/Scripts/Project/sign.jpg" ,HandView) 
            print("Image saved")

        if k == 27:
                break

    cv2.destroyAllWindows()
    cap.release()
    

def show_image():
    image_path = 'C:/Python/Python37/Scripts/Project/sign.jpg'
    image = Image.open(image_path)
    image.show()

# fonts ---> "Courier New", "Segoe UI","Open Sans","Roboto"

LABEL_FONT = ("Segoe UI",12)
BUTTON_FONT = ("Roboto",12)
# main class ---> app 
# Child class of tk.Tk class as it is an app.

class ASL_RecogSys(tk.Tk):
    # self ---> points towards the object so that an object can use variables and methods of a class 
    # Eg: object_name.variable_name or object_name.method(...)
    # *args --> as many arguments/parameters as you like
    # **kwargs --> keyword arguments (dictionaries)
    
    def __init__(self,*args,**kwargs):
        tk.Tk.__init__(self,*args,**kwargs) #Initializing tkinter
        
        
        tk.Tk.wm_title(self,"ASL RecogSys")
        tk.Tk.wm_geometry(self,"1920x1080")

        container = tk.Frame(self) #Frame ---> Window
        container.pack(side = "top", fill = "both", expand = True)
        container.grid_rowconfigure(0,weight = 1) 
        container.grid_columnconfigure(0,weight = 1)
        
        # Store the pages in the application inside a dictionary called frames
        self.frames = dict() 
        
        #Format
        frame_names = (
            
                    StartPage,
                        MainMenu,
                           FingerSpell,
                           TextToASL
            
                      ) #Will be the name of the windows that will be separate classes
        
        for Fclass in frame_names:
            frame = Fclass(container,self)
            self.frames[Fclass] = frame
            frame.grid(row = 0, column = 0, sticky = "nsew") #"nsew" ---> Strech everything to window size

        self.show_frame(StartPage)
 
        
    def show_frame(self,cont):
        frame = self.frames[cont]
        frame.tkraise() # Show page with the help of page_name class as a key


# First window
# Child class of tk.Frame as it is a window
class StartPage(tk.Frame): 
    def __init__(self,parent,controller):
        tk.Frame.__init__(self,parent)
        image1 = tk.PhotoImage(file="C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/StartPage Background.png")
        panel1 = tk.Label(self, image=image1)
        panel1.pack(side='top', fill='both', expand=True)
        panel1.image = image1
        
        s = ttk.Style()
        s.configure('my.TButton', font= BUTTON_FONT)
        button1 = ttk.Button(self,text = "To Main Menu",style = 'my.TButton',
                            command = lambda: controller.show_frame(MainMenu) )
        
        button1.place(relx = 0.5,  #fraction of width for button x_position
                      rely = 0.85, #fraction of height for button y_position
                      width = 230, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        
class MainMenu(tk.Frame): 
    def __init__(self,parent,controller):
        tk.Frame.__init__(self,parent)
        image1 = tk.PhotoImage(file="C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/MainMenu.png")
        panel1 = tk.Label(self, image=image1)
        panel1.pack(side='top', fill='both', expand=True)
        panel1.image = image1
    
        x_pos = 0.5
        y_pos = 0.35
        y_inc = 0.25
        
        s = ttk.Style()
        s.configure('my.TButton', font= BUTTON_FONT)
        button1 = ttk.Button(self,text = "ASL Fingerspelling Mode",style = 'my.TButton',
                    command = lambda: controller.show_frame(FingerSpell) )
        
        button1.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        
        
        y_pos += y_inc
        button2 = ttk.Button(self,text = "Text to ASL Mode",style = 'my.TButton',
                    command = lambda: controller.show_frame(TextToASL) )
        button2.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position

        y_pos += y_inc
        button3 = ttk.Button(self,text = "Back",style = 'my.TButton',
                            command = lambda: controller.show_frame(StartPage) )
        button3.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        
        
class FingerSpell(tk.Frame): 
    def __init__(self,parent,controller):
        tk.Frame.__init__(self,parent)
        image1 = tk.PhotoImage(file="C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/All Window Background.png")
        panel1 = tk.Label(self, image=image1)
        panel1.pack(side='top', fill='both', expand='yes')
        panel1.image = image1

        label = tk.Label(self,text = "ASL Fingerspelling Mode", font = LABEL_FONT)
        label.place(relx = 0.5, rely = 0.1, anchor = tk.CENTER)
        
        x_pos = 0.5
        y_pos = 0.25
        y_inc = 0.25
        s = ttk.Style()
        s.configure('my.TButton', font= BUTTON_FONT)
        
        button1 = ttk.Button(self,text = "Get Camera",style = 'my.TButton',
                            command = capture_and_save )
        button1.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        
        y_pos +=y_inc
        button2 = ttk.Button(self,text = "Display Image",style = 'my.TButton',
                            command = show_image  )
        button2.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        y_pos +=y_inc
        button3 = ttk.Button(self,text = "Back to Main Menu",style = 'my.TButton',
                            command = lambda: controller.show_frame(MainMenu) )
        button3.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        

    
        
class TextToASL(tk.Frame): 
    def __init__(self,parent,controller):
        tk.Frame.__init__(self,parent)
        image1 = tk.PhotoImage(file="C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/All Window Background.png")
        panel1 = tk.Label(self, image=image1)
        panel1.pack(side='top', fill="both", expand='yes')
        panel1.image = image1

        label = tk.Label(self,text = "Text to ASL Mode", font = LABEL_FONT)
        label.place(relx = 0.5, rely = 0.1, anchor = tk.CENTER)
        x_pos = 0.5
        y_pos = 0.25
        y_inc = 0.25
        s = ttk.Style()
        s.configure('my.TButton', font= BUTTON_FONT)
        
        button1 = ttk.Button(self,text = "Back to Main Menu",style = 'my.TButton',
                            command = lambda: controller.show_frame(MainMenu) )
        button1.place(relx = x_pos,  #fraction of width for button x_position
                      rely = y_pos, #fraction of height for button y_position
                      width = 300, #Button width
                      height = 50, #Button Height
                      anchor = tk.CENTER)#Button  position
        
        
app = ASL_RecogSys()
app.iconbitmap("asl_icon.ico")

app.mainloop()
```

```python
from tkinter import *
from PIL import ImageTk
from PIL import Image, ImageDraw, ImageFont
import IPython.display
import os
import cv2
import numpy as np

def main_menu():
    print("To main menu")
    main.create_window()

def ASL_Fingerspell():
    print("ASL fingerspelling and shit")
    windowASL.create_window()
    
def text_to_ASL():
   
    start_page.create_window()
    
def start():
    print('Just started')
    
def stop():
    print('stopped')

    
class Window:
    root = Tk()
    def __init__(self, window_title, window_size, bg_image,button_dict,exit_b,button_Ypos):
        self.title = window_title
        self.size = window_size
        self.background_image = bg_image
        self.buttons_and_func = button_dict
        self.exit = exit_b
        self.button_y = button_Ypos
        
        
    def create_buttons(self):
        button_names = list(self.buttons_and_func.keys())
        y_pos = self.button_y
        
        for i in range(len(button_names)):
            command_given = self.buttons_and_func[button_names[i]]
            y_pos = y_pos + 0.25
            button = Button(master = self.root,
                            activebackground = 'Blue',
                            text = button_names[i], 
                            font = 'Courier', 
                            height= 2, 
                            width=25, 
                            relief=RAISED,
                            bd='5',
                            command = command_given)
            button.place(relx = 0.5, rely = y_pos, anchor = CENTER) 
            
            
        #Exit Button
        if self.exit == True:
            button = Button(master = self.root,
                            activebackground = 'Red',
                            text = "Exit", 
                            font = 'Courier', 
                            height= 2, 
                            width=25, 
                            relief=RAISED,
                            bd='5',
                            command = self.root.destroy)
            button.place(relx = 0.5, rely = 0.25 * (len(button_names)+1), anchor = CENTER)   
           
            
        
    def create_window(self):
        self.root.geometry(self.size)  # For a fixed size of window
        self.root.title(self.title )

        #To set the background Image
        image = Image.open(self.background_image)
        background = ImageTk.PhotoImage(image)
        background_label = Label(image = background)
        background_label.place(x=0, y=0, relwidth=1, relheight=1) 
        
    
        self.create_buttons()
     
        self.root.mainloop()
        

# Window 1
title0 = 'American Sign Language Recognition'
size0 = "1280x868"
bg_image0 = "C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/bg-0.png"
button_dict0 = {"To Main Menu": main_menu}

start_page = Window(window_title = title0,
              window_size = size0,
              bg_image = bg_image0,
              button_dict = button_dict0,
              exit_b = False,
              button_Ypos = 0.60)

        
# Window 2
title1 = 'Main Menu'
size1 = "1280x868"
bg_image1 = "C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/bg-1-2.png"
button_dict1 = {"ASL Finger Spelling":ASL_Fingerspell, "Text to ASL": text_to_ASL}

main = Window(window_title = title1,
              window_size = size1,
              bg_image = bg_image1,
              button_dict = button_dict1,
              exit_b = True,
              button_Ypos = 0)




def go_back():
    main.create_window()
    try:
        windowASL.destroy()
    except:
        pass

# Window 3
title2 = 'ASL Finger Spelling'
size2 = "1280x868"
bg_image2 = "C:/Python/Python37/Scripts/Project/GitHub/GUI-for-ASLrecog/bg-1-2.png"
button_dict2 = {"Start":start, "Stop": stop, "Back":go_back}

windowASL = Window(window_title = title2,
                   window_size = size2,
                   bg_image = bg_image2,
                   button_dict = button_dict2,
                   exit_b = False,
                   button_Ypos = 0)


start_page.create_window()
```
